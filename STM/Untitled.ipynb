```python
import numpy as np
import matplotlib.pyplot as mpl
from scipy import stats
```

```python
#stm
data_1=np.loadtxt("new 2.txt", comments="#", unpack=False, skiprows=1)
length_0, voltage_0, sigma_v_0=np.hsplit(data_1,3)

length= length_0[:,0]
voltage=voltage_0[:,0]
sigma_v=sigma_v_0[:,0]
vol=voltage/0.335

print(length)
print(voltage)
print(sigma_v)
print(vol)
print(np.average(voltage))
print(np.std(voltage, dtype=np.float64))
print(stats.sem(voltage, axis=None, ddof=0))
print(np.mean(voltage, dtype=np.float64))

```

```python
mpl.errorbar(length, voltage, yerr=sigma_v, fmt="r.",markersize=10, label='data')
mpl.xlabel('Number of measurements(N)')
mpl.ylabel('Height in nm (H)')

mpl.xlim(0,)
mpl.ylim(0,2.0)

mpl.legend()
```

```python
mpl.errorbar(length, vol, yerr=sigma_v, fmt="r.",markersize=10, label='data')
mpl.xlabel('Number of measurements(N)')
mpl.ylabel('Normalized Height in nm (H)')

mpl.xlim(0,)
mpl.ylim(0,3.0)

mpl.legend()
```
